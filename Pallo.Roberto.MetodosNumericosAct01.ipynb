**4.-**Sea la función f(x) =$\sqrt{x^{2}+2x+1}$ calcule $f(2.045)$ con un error de $0.00005$ (utilice una serie de
Taylor).

Sea la formula de la serie de Taylor igual a:
<h3>$p(x)$=$f(x)$+$f^{´}(x-x_0)$ + $\frac{ f^{''}(x_0)(x-x_0)^{2}}{2}$ + $\frac{ f^{'''}(x_0)(x-x_0)^{3}}{3!}$ + $\frac{ f^{''''}(x_0)(x-x_0)^{4}}{4!}$ +...+ $\sum_{n=0}^{\infty}$=$\frac{ f^{n}(x_0)(x-x_0)^{n}}{n!}$</h3>
Entonces vamos a calcular las  3 primeras derivadas de
f(x) =$\sqrt{x^{2}+2x+1}$
<h3>$f^{´}$=$\frac{1}{2\sqrt{x^{2}+2x+1}}\cdot(x^{2}+2x+1)^{'};$</h3>
<h3>$f^{´}$=$\frac{(2x+2)}{2\sqrt{x^{2}+2x+1}};$</h3>
<h3>$f^{''}$=$\frac{1}{2}\cdot\frac{(2x+2)^{´}\sqrt{x^{2}+2x+1}-(2x+2)(\sqrt{x^{2}+2x+1})^{´}}{(\sqrt{x^{2}+2x+1})^{2}};$</h3>
<h3>$f^{''}$=$\frac{1}{2}\cdot\frac{2\sqrt{x^{2}+2x+1}-(2x+2)}{(x+1)^{2}} = 0;$</h3>


Nos podemos dar cuenta que a partir $f^{''}$ el resultado resulta ser cero y la derivada de cero es cero por lo tanto no hace falta calcular más derivadas.
Evaluamos f(2.045) en:
<h3>$f(2.045) =\sqrt{(2.045)^{2}+2(2.045)+1};$</h3>
<h3>$f(2.045) =\sqrt{(4.18203)+(4.09000)+1};$</h3>
<h3>$f(2.045) ={3.04500}$</h3>
<h3></h3>Evaluamos f(2.045) en:

<h3>$f(2.045)^{´}$=$\frac{2(2.045)+2}{2\sqrt{(2.045)^{2}+2(2.045)+1}};$</h3>
<h3>$f(2.045)^{´}$=$\frac{6,09000}{6,09000};$</h3>

Reemplazamos los valores obtenidos en la serie de Taylor 
<h3>$p(x)= 3.04500 + 1\cdot(x-2.045) + 0 $</h3>

6.Diseñe un código que encuentre el $ Sen(\frac{pi}{3})$ a través del desarrollo de Taylor, truncar cuando n = 50.

A continuación recordaremos la función sen(x) en forma de la sumatoria con respecto a k:
<h3> $sen(x)=\sum_{k=0}^{k=n-1}={(-1)}^{k}\cdot\frac {x^{(2k + 1)}}{(2k+1)!}$</h3>


```python
import numpy as np
import math
# Datos:declaración de Variables
x=np.pi/3
#Número de veces a truncar un numero:
#Definimos una variable que sea la que guarde la suma de la serie: 
n=int(2)
sumatoria=0
#Ciclo para repetir y acumular el valor de las sumas de la serie de Taylor
for k in range(1,n+1):
    sumatoria=sumatoria + (-1)**(k+1)*x**(2*k+1)/np.math.factorial(2*k+1)

```
